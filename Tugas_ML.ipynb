```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
dataku = pd.read_csv("/content/german.csv")
print(dataku)
```

```python
#mengubah data ke numerik
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for col in dataku.columns.values:
  #encoding pada variabel kategori
  if dataku[col].dtypes=='object' or dataku[col].dtype=='bool':
    data = dataku[col].append(dataku[col])
    le.fit(data.values)
    dataku[col] = le.transform(dataku[col])
dataku.head(12)
```

```python
print("Ukuran :",dataku.shape)
```

```python
x = dataku[dataku.columns[:-1]]
y = dataku[dataku.columns[-1]]
print("\nFeatures matrix:\n",x)
print("\nResponse vector:\n",y)
```

```python
print("\nFeatures:",dataku.columns)
```

```python
from sklearn.model_selection import train_test_split
# from pandas import DataFrame
# from pandas.core.describe import DataFrameDescriber
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=1)

print(x_train)
print(x_train.shape)

print(x_test)
print(x_test.shape)

print(y_train)
print(y_train.shape)

print(y_test)
print(y_test.shape)
```

```python
from sklearn.metrics import euclidean_distances
from sklearn.neighbors import KNeighborsClassifier

#euclidean distances
knn = KNeighborsClassifier(n_neighbors=2)

#data latih
knn.fit(x_train,y_train)
```

```python
y_pred = knn.predict(x_test)
from sklearn import metrics
print("Akurasi model KNN:",metrics.accuracy_score(y_test,y_pred))
```
